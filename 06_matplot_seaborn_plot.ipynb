```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# Create a 2x2 confusion matrix
cm_2x2 = np.array([[50, 10],  # True Positive, False Negative
                   [5, 35]])  # False Positive, True Negative

# Display the 2x2 confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm_2x2, display_labels=["Class 1", "Class 2"])
disp.plot(cmap="Blues", xticks_rotation=45)

plt.title("2x2 Confusion Matrix")
plt.show()
```

```python
# Create a 4x4 confusion matrix
cm_4x4 = np.array([[40, 5, 8, 2],    # Class 1 predictions
                   [4, 50, 6, 3],    # Class 2 predictions
                   [5, 7, 45, 1],    # Class 3 predictions
                   [2, 4, 3, 60]])   # Class 4 predictions

# Display the 4x4 confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm_4x4, display_labels=["Class 1", "Class 2", "Class 3", "Class 4"])
disp.plot(cmap="Blues", xticks_rotation=45)

plt.title("4x4 Confusion Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay

# Set global font size
plt.rcParams.update({'font.size': 30})  # Change font size globally

# Example: 3x3 confusion matrix
cm_3x3 = np.array([[30, 5, 10],
                   [7, 40, 8],
                   [6, 4, 50]])
disp = ConfusionMatrixDisplay(confusion_matrix=cm_3x3, display_labels=["Class 1", "Class 2", "Class 3"])
disp.plot(cmap="Blues", xticks_rotation=45)

plt.title("Confusion Matrix (Font Size Changed)")
plt.show()
```

```python
# Example: Manually adjust specific font sizes
disp = ConfusionMatrixDisplay(confusion_matrix=cm_3x3, display_labels=["Class 1", "Class 2", "Class 3"])
fig, ax = plt.subplots(figsize=(8, 6))
disp.plot(ax=ax, cmap="Blues", xticks_rotation=45)

# Change title font size
ax.set_title("Confusion Matrix (Custom Font Sizes)", fontsize=20)

# Change x-axis and y-axis label font size
ax.set_xlabel("Predicted Labels", fontsize=16)
ax.set_ylabel("True Labels", fontsize=16)

# Change tick label font size
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example: 3x3 confusion matrix
cm = np.array([[30, 5, 10],
               [7, 40, 8],
               [6, 4, 50]])

# Class labels
labels = ["Class 1", "Class 2", "Class 3"]

# Plot the confusion matrix
fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(cm, cmap="Blues")  # Plot matrix
fig.colorbar(cax)

# Set axis labels
ax.set_xticks(range(len(labels)))
ax.set_yticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha="right", fontsize=12)
ax.set_yticklabels(labels, fontsize=12)
ax.set_xlabel("Predicted Labels", fontsize=14)
ax.set_ylabel("True Labels", fontsize=14)

# Add custom annotations for the values
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        color = "white" if cm[i, j] > cm.max() / 2 else "black"
        ax.text(j, i, f"{cm[i, j]}", ha="center", va="center", color=color, fontsize=16)

# Set title
ax.set_title("Custom Confusion Matrix", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Example: 3x3 confusion matrix
cm = np.array([[30, 5, 10],
               [7, 40, 8],
               [6, 4, 50]])

# Class labels
labels = ["Class 1", "Class 2", "Class 3"]

# Create the confusion matrix heatmap
fig = go.Figure(
    data=go.Heatmap(
        z=cm,
        x=labels,  # Predicted labels
        y=labels,  # True labels
        colorscale="Blues",
        showscale=True,
        text=cm,  # Values to display inside the heatmap
        texttemplate="%{text}",  # Display raw numbers
        textfont={"size": 16}  # Customize font size for annotations
    )
)

# Customize layout
fig.update_layout(
    title="Confusion Matrix (Interactive)",
    xaxis_title="Predicted Labels",
    yaxis_title="True Labels",
    xaxis=dict(tickangle=45),
    font=dict(size=14),  # General font size
    height=500,
    width=600,
)

# Show the plot
fig.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Example: 3x3 confusion matrix
cm = np.array([[30, 5, 10],
               [7, 40, 8],
               [6, 4, 50]])

# Class labels
labels = ["Class 1", "Class 2", "Class 3"]

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    cm,
    annot=True,            # Annotate cells with numbers
    fmt="d",               # Format numbers as integers
    cmap="Blues",          # Color map
    xticklabels=labels,    # X-axis labels
    yticklabels=labels,    # Y-axis labels
    cbar=True,             # Add a color bar
    annot_kws={"size": 16} # Font size for annotations
)

# Add labels and title
plt.title("Confusion Matrix (Seaborn)", fontsize=18)
plt.xlabel("Predicted Labels", fontsize=14)
plt.ylabel("True Labels", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
# Data
data = {
    "Class": [5, 0, 4, 1, 9, 2, 3, 6, 7, 8],
    "Images": [5421, 5923, 5842, 6742, 5949, 5958, 6131, 5918, 6265, 5851],
}

df = pd.DataFrame(data)

# Seaborn barplot
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x="Class", y="Images", palette="Blues_d")

# Add title and labels
plt.title("Class Balance Distribution (Seaborn)", fontsize=7)
plt.xlabel("Class", fontsize=7)
plt.ylabel("Number of Images", fontsize=7)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Matplotlib barplot
plt.figure(figsize=(10, 6))
plt.bar(df["Class"], df["Images"], color="skyblue")

# Add title and labels
plt.title("Class Balance Distribution (Matplotlib)", fontsize=16)
plt.xlabel("Class", fontsize=14)
plt.ylabel("Number of Images", fontsize=14)
plt.xticks(df["Class"], fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Normalize the data for the colormap
norm = plt.Normalize(min(df['Images']), max(df['Images']))
colors = cm.Blues(norm(df['Images']))  # Use the "Blues" colormap

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Class'], df['Images'], color=colors)


# Add title and labels
plt.title("Class Balance Distribution (Matplotlib with Colormap)", fontsize=16)
plt.xlabel("Class", fontsize=14)
plt.ylabel("Number of Images", fontsize=14)

# Add gridlines
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go
import seaborn as sns

# Data
classes = [5, 0, 4, 1, 9, 2, 3, 6, 7, 8]
images = [5421, 5923, 5842, 6742, 5949, 5958, 6131, 5918, 6265, 5851]

# Generate a dynamic color palette using Seaborn
colors = sns.color_palette("Blues_d", len(classes)).as_hex()  # Convert to hex for Plotly

# Create the Plotly bar chart
fig = go.Figure(data=go.Bar(
    x=classes,  # Class labels
    y=images,  # Image counts
    text=images,  # Annotate bars with counts
    textposition="auto",  # Place text on top of the bars
    marker=dict(color=colors),  # Use the dynamic color palette
))

# Customize layout
fig.update_layout(
    title="Class Balance Distribution (Plotly with Dynamic Colors)",
    xaxis_title="Class",
    yaxis_title="Number of Images",
    xaxis=dict(tickmode="linear"),
    height=500,
    width=700
)

# Show the plot
fig.show()
```
